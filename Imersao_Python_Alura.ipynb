```python
dfPrincipal = dfPrincipal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
dfPrincipal
```

```python
dfPrincipal = dfPrincipal.rename(columns={'Último (R$)':'valorFinal', 'Var. Dia (%)':'varDiaPercent'}).copy()
dfPrincipal
```

Criando uma nova coluna:

```python
dfPrincipal['varPercent'] = dfPrincipal['varDiaPercent'] / 100
dfPrincipal
```

```python
dfPrincipal['valorInicial'] = dfPrincipal['valorFinal'] / (dfPrincipal['varPercent'] + 1)
dfPrincipal
```

Mergeando duas colunas:

```python
 dfPrincipal = dfPrincipal.merge(dfTotalAcoes, left_on='Ativo', right_on='Código', how='left')
 dfPrincipal
```

Excluindo uma coluna específica repetida:

```python
dfPrincipal = dfPrincipal.drop(columns=['Código'])
dfPrincipal
```

```python
dfPrincipal['variacaoDinheiro'] = (dfPrincipal['valorFinal'] - dfPrincipal['valorInicial']) * dfPrincipal['Qtde. Teórica']
dfPrincipal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
dfPrincipal
```

```python
dfPrincipal['Qtde. Teórica'] = dfPrincipal['Qtde. Teórica'].astype(int)
dfPrincipal
```

```python
dfPrincipal = dfPrincipal.rename(columns={'Qtde. Teórica':'qtdTeorica'}).copy()
dfPrincipal
```

```python
def verificar_tendencia(valor):
    return "Subiu" if valor > 0 else ("Desceu" if valor < 0 else "Estável")
```

```python
dfPrincipal['resultado'] = dfPrincipal['variacaoDinheiro'].apply(verificar_tendencia)
dfPrincipal
```

```python
# ou poderiamos aplicar a fórmula:
# df_principal['resultado'] = df_principal['variacaoDinheiro'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
```

```python
 dfPrincipal = dfPrincipal.merge(dfTicker, left_on='Ativo', right_on='Ticker', how='left')
 dfPrincipal = dfPrincipal.drop(columns=['Ticker'])
 dfPrincipal
```

```python
 dfPrincipal = dfPrincipal.merge(dfChatGPT, left_on='Nome', right_on='Nome da empresa', how='left')
 dfPrincipal = dfPrincipal.drop(columns=['Nome da empresa'])
 dfPrincipal
```

```python
dfPrincipal = dfPrincipal.drop_duplicates()
dfPrincipal
```

```python
def verificar_categoria_idades(valor):
    return "Mais de 100" if valor > 100 else ("Menos de 50" if valor < 50 else "Entre 50 e 100")
```

```python
dfPrincipal['faixaIdade'] = dfPrincipal['Idade (anos)'].apply(verificar_categoria_idades)
dfPrincipal
```

# Análises

```python
# Calcular o maior valor
maior_valor = dfPrincipal['variacaoDinheiro'].max()

# Calcular o menor valor
menor_valor = dfPrincipal['variacaoDinheiro'].min()

# Calcular a média
media = dfPrincipal['variacaoDinheiro'].mean()

# Calcular a média de quem subiu
media_subiu = dfPrincipal[dfPrincipal['resultado'] == 'Subiu']['variacaoDinheiro'].mean()

# Calcular a média de quem desceu
media_desceu = dfPrincipal[dfPrincipal['resultado'] == 'Desceu']['variacaoDinheiro'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior_valor:,.2f}")
print(f"Menor\tR$ {menor_valor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
dfPrincipalSubiu = dfPrincipal[dfPrincipal['resultado'] == 'Subiu']
```

```python
dfPrincipalSubiu
```

```python
dfSegmento = dfPrincipalSubiu.groupby('Segmento')['variacaoDinheiro'].sum().reset_index()
dfSegmento
```

```python
dfSaldo = dfPrincipal.groupby('resultado')['variacaoDinheiro'].sum().reset_index()
dfSaldo
```

```python
import plotly.express as px
```

```python
fig = px.bar(dfSaldo, x='resultado', y='variacaoDinheiro', text='variacaoDinheiro', text_auto=True, title='Variação em R$ versus Resultado')
fig.update_layout(width=800, height=300, margin=dict(l=50, r=50, t=50, b=50))
fig.update_layout(title_font=dict(family="Arial", size=18, color="blue"))
fig.update_traces(texttemplate='%{text:.2f}')
fig.show()
```

```python
figsegmentos = px.pie(dfSegmento, values='variacaoDinheiro', names='Segmento', title='Segmento versus Variação R$')
fig.update_traces(textposition='inside')
figsegmentos.update_layout(uniformtext_minsize=12, uniformtext_mode='hide')
figsegmentos.show()
```

```python
dfFaixaIdade = dfPrincipal.groupby('faixaIdade')['variacaoDinheiro'].sum().reset_index()
dfFaixaIdade
```

```python
figFaixaIdade = px.bar(dfFaixaIdade, x='faixaIdade', y='variacaoDinheiro', text='variacaoDinheiro', title='Faixa de Idade versus Resultado')
figFaixaIdade.update_layout(width=800, height=400, margin=dict(l=50, r=50, t=50, b=50), legend=dict(y=1.1, x=0.5))
figFaixaIdade.update_layout(title_font=dict(family="Arial", size=12, color="blue"))
figFaixaIdade.update_traces(texttemplate='%{text:.2f}')
figFaixaIdade.show()
```

```python
valor_procurado1 = 'Entre 50 e 100'
contagem1 = dfPrincipal['faixaIdade'].eq(valor_procurado1).sum()
valor_procurado2 = 'Mais de 100'
contagem2 = dfPrincipal['faixaIdade'].eq(valor_procurado2).sum()
valor_procurado3 = 'Menos de 50'
contagem3 = dfPrincipal['faixaIdade'].eq(valor_procurado3).sum()

dfFaixaIdade.at[0, 'qtdEmpresas'] = contagem1
dfFaixaIdade.at[1, 'qtdEmpresas'] = contagem2
dfFaixaIdade.at[2, 'qtdEmpresas'] = contagem3
```

```python
dfFaixaIdade
```

```python
figfaixaIdadePie  = px.pie(dfFaixaIdade, values='qtdEmpresas', names='faixaIdade', title='Quantidade de Empresas %')
figfaixaIdadePie.update_traces(textposition='inside')
figfaixaIdadePie.update_layout(uniformtext_minsize=12, uniformtext_mode='hide')
figfaixaIdadePie.show()
```
